#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Ans:
It is a 2D plotting library that provides a variety of tools for creating static, animated, and interactive visualizations in Python. It is widely used in scientific computing, data analysis, and data visualization tasks.

Matplotlib can be used for a variety of tasks such as creating line plots, scatter plots, bar plots, histograms, pie charts, and more.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

- Line plot: A line plot is a plot of data points connected by straight lines. It is useful for visualizing trends in data over time or space.

- Scatter plot: A scatter plot is a plot of points that shows the relationship between two variables. It is useful for visualizing patterns and trends in data.

- Bar plot: A bar plot is a plot of bars that shows the frequency or proportion of data in different categories. It is useful for comparing data across different categories.

- Histogram: A histogram is a plot of the distribution of a single variable. It is useful for visualizing the shape of a data distribution and identifying outliers.

- Pie chart: A pie chart is a plot of a data set that shows the proportion of each category in a whole. It is useful for showing the relative sizes of different categories.

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
- plot a scatter plot.
- import numpy as np
- np.random.seed(3)
- x = 3 + np.random.normal(0, 2, 50)
- y = 3 + np.random.normal(0, 2, len(x))
- Note: Also add title, xlabel, and ylabel to the plot. **

#### Ans:
A scatter plot, also known as a scatter chart, is a type of plot used to display the relationship between two variables. It is a graph in which the values of two variables are plotted as points on a two-dimensional coordinate system, with one variable on the x-axis and the other on the y-axis.

```python
# Importing numpy and matplotlib

import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
```

```python
x = 3 + np.random.normal(0,2,50)
```

```python
y = 3 + np.random.normal(0, 2, len(x))
```

```python
# plot the x,y 
plt.scatter(x,y)

# add title
plt.title("x vs y")

# add xlabel
plt.xlabel('x-axis')

#add ylabel
plt.ylabel("y-axis")
```

##### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
- import numpy as np
- For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
- For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
- For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
- For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

#### Ans:
The subplot() function is used to create multiple plots in a single figure. It allows us to divide a figure into a grid of subplots, where each subplot can display a different plot or visualization. 

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot the data in each subplot
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Set the titles for each subplot
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Add a common x-label and y-label
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.04, 0.5, 'Y-axis', va='center', rotation='vertical')

# Adjust the spacing between subplots
fig.tight_layout()

# Display the figure
plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
- import numpy as np
- company = np.array(["Apple", "Microsoft", "Google", "AMD"])
- profit = np.array([3000, 8000, 1000, 10000])

#### Ans:
A bar plot is a type of chart or graph that represents categorical data with rectangular bars. The length or height of each bar corresponds to the value of the category or group it represents.

Bar plots are useful for visualizing data that can be divided into discrete categories, such as counts or frequencies of categorical variables. They can also be used to compare data across multiple groups or subgroups.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(['Apple', 'Microsoft', 'Google', 'AMD'])
profit = np.array([3000, 8000, 1000, 10000])

#creat bar plot
plt.bar(company, profit)
plt.title("Bar graph")
plt.xlabel("Companies")
plt.ylabel("profit")
```

```python
# create horizontal bar
plt.barh(company, profit)

plt.xlabel("profit")
plt.ylabel("Companies")
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
- box1 = np.random.normal(100, 10, 200)
- box2 = np.random.normal(90, 20, 200)

#### Ans:
A box plot, also known as a box and whisker plot, is a type of chart or graph used to visualize the distribution of a dataset. It shows the median, quartiles, and outliers of a dataset.

Box plots are useful for identifying the skewness, spread, and outliers of a dataset. They can also be used to compare the distribution of multiple datasets or groups.

```python
import numpy as np
import matplotlib.pyplot as  plt


# create dataset:
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90,20,200)
data = [box1, box2]

#create box plot
plt.boxplot(data)
plt.ylabel('Value')
plt.title('Box Plot')

plt.show()
```
