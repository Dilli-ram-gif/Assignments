#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

## Ans:
Flask is a lightweight web application framework written in Python that provides tools and libraries for building web applications. Flask is a microframework, which means that it is minimalistic and does not require any particular tools or libraries to be installed.

Here are some advantages of Flask framework:

1. Lightweight: Flask is a lightweight framework, making it easy to use and understand for beginners and experienced developers alike.

2. Easy to learn: Flask has a simple and easy-to-understand API that makes it easy to learn for new developers.

3. Highly customizable: Flask is highly customizable and provides flexibility for developers to design and implement their applications the way they want.

4. Modular: Flask is a modular framework that allows developers to choose which components they want to use.

5. Extensible: Flask is highly extensible, with a large number of third-party extensions and plugins available.

6. Large community: Flask has a large and active community, with plenty of resources available to help developers learn and troubleshoot issues.

#### Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
# Creating simple Flask application to display Hello world!

from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run()
```

#### Q3. What is App routing in Flask? Why do we use app routes?

Ans: App routing refers to the process of binding URL routes to view functions in a Flask application. It allows us to map specific URLs (or routes) to corresponding functions that generate HTTP responses.

For example, we can map the URL /home to a function that returns the HTML content for the homepage of our application. This makes it easy to create dynamic web pages with content that changes based on the URL requested by the user.

We use app routes in Flask to define the behavior of our web application. By mapping URLs to specific functions, we can control what content is displayed to the user based on their requests. App routes allow us to create dynamic web pages that respond to user input, such as filling out a form or clicking a link.

#### Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details: Company Name: ABC Corporation , Location: India Contact,  Detail: 999-999-9999

```python
# Creating route and displaying welcome message

from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_info():
    return '''Company Name: ABC Corporation
              Location: India
              Contact Detail: 999-999-9999'''

if __name__ == '__main__':
    app.run()
```

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

Ans: The url_for() function is used for URL building.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page'

@app.route('/profile/<username>')
def profile(username):
    return f'This is the profile page of {username}'

@app.route('/user')
def user():
    # generate the URL for the profile() function with username parameter as ''Dilli
    profile_url = url_for('profile', username='John')
    return f'The URL for John\'s profile is {profile_url}'

if __name__ == '__main__':
    app.run()
```
